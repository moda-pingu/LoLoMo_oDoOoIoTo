```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
```

```python
# api_key 값을 .env 파일에서 불러오기
# : load_dotenv() 만 해도 된다 
# : ChatOpenAI(...) 생성 시 내부 OpenAI SDK가 OPENAI_API_KEY 환경변수를 자동 조회해서 요청 헤더에 사용

load_dotenv()
MYLLM = ChatOpenAI(
    model="gpt-4o",
    # api_key=os.getenv('OPENAI_API_KEY'), # 없어도 된다: load_dotenv() 로 .env 파일에서 불러온 OPENAI_API_KEY 환경변수 사용 
    # ChatOpenAI(...) 생성 시 내부 OpenAI SDK가 OPENAI_API_KEY 환경변수를 자동 조회해서 요청 헤더에 사용
)

from langchain_core.messages import HumanMessage
response = MYLLM.invoke([HumanMessage(content="안녕?")])
print(response.content)

# openai 라이브러리로는 아래와 같이 작성
# response = openai.ChatCompletion.create(
#     model="gpt-4o",
#     messages=[{"role": "user", "content": "안녕?"}]
# )
```

```python
# Langchain 의 메세지 타입
# [system]    SystemMessage: 시스템 메시지 (예: 모델에게 역할이나 행동 지침을 제공)
# [user]      HumanMessage: 사용자 메시지 (예: 모델에게 질문이나 요청을 전달)
# [assistant] AIMessage: 모델의 응답 메시지 (예: 모델이 생성한 답변이나 결과를 나타냄)
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

MYLLM = ChatOpenAI(
    model="gpt-4o",
    temperature=0.5
)

messages = [
    SystemMessage(content="너는 친절한 조수야."), # 시스템 메시지: 모델에게 역할과 행동 지침 제공
    HumanMessage(content="안녕?"), # 사용자 메시지: 모델에게 질문이나 요청 전달
]
```
