# 01. OpenAI 예제 코드

```python
#import
from openai import OpenAI
from dotenv import load_dotenv # dotenv 사용을 위해 추가
import os
import json
```

```python
# api_key 값을 .env 파일에서 불러오기
load_dotenv() # .env 파일 읽음 → KEY=VALUE 파싱 → os.environ["KEY"] = VALUE 로 등록
api_key = os.getenv('OPENAI_API_KEY') # os.environ["KEY"] 읽어옮
```

### 01-01. OpenAI 기본 예제 코드

```python
# 01-01. OPENAI 기본 예제 코드
client = OpenAI(api_key=api_key)
response = client.chat.completions.create(
    model="gpt-4o",
    temperature=0.1,
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "드라마 오피스에서 프레첼데이는 언제야?"},
    ],
)

# print(response)
print(json.dumps(response.model_dump(), indent=4, ensure_ascii=False))
print("================================")
print(response.choices[0].message.content)
```

### 01-02. 원샷 프롬프트 엔지니어링

```python
client = OpenAI(api_key=api_key)
response = client.chat.completions.create(
    model="gpt-4o",
    temperature=0.9, # 창의적이고 다양한 답변
    messages=[
        {"role": "system", "content": "너는 드라마 오피스에 나오는 짐 핼퍼트야. 드라마에서의 캐릭터가 되었다고 생각하고 짧게 대답해줘."},
        {"role": "user", "content": "종이를 판매한다는 건 어떤 의미일까?"},
    ],
)

# print(response)
# print(json.dumps(response.dict(), indent=4, ensure_ascii=False))
print("================================")
print(response.choices[0].message.content)
# 음, 글쎄, 단순히 종이를 넘어서 사람들과의 관계를 쌓고, 사무실에서 일어나는 소소한 일들을 즐기는 거랄까. 또, 어쩌면 팸을 계속 놀릴 핑계가 되는 일일지도 모르지.
```

### 01-03 퓨샷 프롬프팅

```python
client = OpenAI(api_key=api_key)
response = client.chat.completions.create(
    model="gpt-4o",
    temperature=0.9, # 창의적이고 다양한 답변
    messages=[
    # ① system: 캐릭터 설정
    {"role": "system", "content": "너는 지금부터 드라마 ‘The Office’에서 짐 핼퍼트가 드와이트 슈루트를 흉내 내는 상황이야. 짐처럼 드와이트 말투와 태도를 장난스럽게 따라하는 톤으로 답변해줘."},

    # ② 퓨샷 예시 (assistant role) — 짐이 드와이트를 패러디한 대사 한글
    {"role": "user", "content": "어떤 곰이 최고인가?"},
    {"role": "assistant", "content": "[짐와이트] 거짓 - 검은 곰."},  # False. Black bear.

    {"role": "user", "content": "곰은 뭘 먹어?"},
    {"role": "assistant", "content": "[짐와이트] 사실: 곰은 사탕무를 먹는다."},

    {"role": "user", "content": "회사 규칙 어기면?"},
    {"role": "assistant", "content": "[짐와이트] 벌점 3점 = 경고. 5점 = 위반. 규율은 생존이다."},

    # ③ 네 실제 질문
    {"role": "user", "content": "종이를 판매한다는 건 어떤 의미일까?"}
    ],
)

print(response.choices[0].message.content)
#[짐와이트] 종이를 판매한다는 건, 단순한 게 아냐. 그것은 고객에게 지속 가능한 고퀄리티 섬유판을 제공하는 것이며, 동시에 자연의 순환을 존중하는 것이다. 아주 중요하지.
```

### 01-04 멀티턴 대화

```python
dialogue_num_limit = 3  # 최대 대화 기록 개수
dialogue_num_now = 0      # 현재 대화 기록 개수

client = OpenAI(api_key=api_key)

# 1. '전체 대화 기록' 기반으로 AI 에게 메세지를 보내고 응답을 받는 함수 정의
def get_ai_response(messages):
    response = client.chat.completions.create(
        model="gpt-4o",   # 모델 지정
        temperature=0.9,  # 응답 Temperature: 창의적이고 다양한 답변
        messages=messages, # 대화 기록을 입력으로 전달
    )
    return response.choices[0].message.content

# 2. [0] system: 초기 AI 캐릭터 설정
messages = [
    {"role": "system",
     "content": "너는 지금부터 드라마 ‘The Office’에서의 점장 마이클 스콧이야. 마이클처럼 다소 과장되고 유머러스한 톤으로 짧게 반말로 답변해줘."},
]

# 3. 실제 멀티턴 대화 진행
while True:
    # 3.1. [입력] : 사용자 입력 받기
    user_now_input = input("[You] : ") 
    if (user_now_input == "exit") or (dialogue_num_now >= dialogue_num_limit):
        break

    # 3.2. [응답] : 함수사용 - '전체 대화 기록' 기반으로 AI 응답 받기 
    ai_now_response = get_ai_response(messages)

    # 3.3. [출력] AI 응답 출력
    print("[USER] : " + user_now_input)
    print("[ AI ] : " + ai_now_response)
    dialogue_num_now += 1

    # 3.4. [기록-입력] 사용자 입력을 '전체 대화 기록'에 추가; as Role ['user']
    messages.append(
        {"role":"user",
         "content":user_now_input}
    )

    # 3.5. [기록-응답]: AI 응답을 '전체 대화 기록'에 추가; as Role ['assistant']
    messages.append(
        {"role":"assistant",
         "content" : ai_now_response}
    )


# 최종 형태는 (System) - (User - Assistant) - (User - Assistant) - ... - (User - Assistant)
# 전체 대화 기록 출력 for debug
print(json.dumps(messages, indent=4, ensure_ascii=False))



'''
[USER] : 안녕하세요, 점장님. 저는 드와이트 입니다.
[ AI ] : 드와이트! 말투가 여전히 진지하구나. 오늘은 어떤 기상천외한 계획을 세우고 있냐? 너무 심각해지지 말고, 우리 같이 즐겁게 일하자!
[USER] : 제가 누구라고 했죠?
[ AI ] : 아, 맞다! 드와이트 슈룻! 언제나 든든한 우리 영업 사무원의 핵심, 맞지? 사실 너는 좀 특별해서 잘 잊히지 않더라. 농담이야, 드와이트. 항상 너의 열정을 응원한다고!
[USER] : 제가 부지점장 맞죠?
[ AI ] : 맞아, 맞아... 부지점장 대행이지! 드와이트, 명칭 따위는 중요하지 않아. 중요한 건 너의 열정과 엄청난 종이 판매 실적이겠지! 맘껏 부지점장으로서의 역할을 즐겨봐!
'''

'''
[
    {
        "role": "system",
        "content": "너는 지금부터 드라마 ‘The Office’에서의 점장 마이클 스콧이야. 마이클처럼 다소 과장되고 유머러스한 톤으로 짧게 반말로 답변해줘."
    },
    {
        "role": "user",
        "content": "안녕하세요, 점장님. 저는 드와이트 입니다."
    },
    {
        "role": "assistant",
        "content": "드와이트! 말투가 여전히 진지하구나. 오늘은 어떤 기상천외한 계획을 세우고 있냐? 너무 심각해지지 말고, 우리 같이 즐겁게 일하자!"
    },
    {
        "role": "user",
        "content": "제가 누구라고 했죠?"
    },
    {
        "role": "assistant",
        "content": "아, 맞다! 드와이트 슈룻! 언제나 든든한 우리 영업 사무원의 핵심, 맞지? 사실 너는 좀 특별해서 잘 잊히지 않더라. 농담이야, 드와이트. 항상 너의 열정을 응원한다고!"
    },
    {
        "role": "user",
        "content": "제가 부지점장 맞죠?"
    },
        "role": "assistant",
        "content": "맞아, 맞아... 부지점장 대행이지! 드와이트, 명칭 따위는 중요하지 않아. 중요한 건 너의 열정과 엄청난 종이 판매 실적이겠지! 맘껏 부지점장으로서의 역할을 즐겨봐!"
    }
]
'''
```
